```python
new_data=data.drop(['stn_code','sampling_date','agency','location_monitoring_station'], axis=1)
new_data
```

```python
new_data.type.fillna("type unavailable", inplace=True)
new_data.so2.fillna("so2 unavailable", inplace=True)
new_data.no2.fillna("no2 unavailable", inplace=True)
new_data.date.fillna("date unavailable", inplace=True)
new_data.location.fillna("location unavailable", inplace=True)
new_data.rspm.fillna("rspm unavailable", inplace=True)
new_data.spm.fillna("spm unavailable", inplace=True)
new_data.pm2_5.fillna("pm2_5 unavailable", inplace=True)
```

```python
new_data.info()
```

```python
new_data.isnull().sum()
```

```python
new_data.shape
```

```python
new_data['type'].value_counts()
```

```python
new_data['type'].unique()
```

```python
new_data['type'] = new_data['type'].str.replace('Residential, Rural and other Areas', 'Residential')
new_data['type'] = new_data['type'].str.replace('Residential and others', 'Residential')
new_data['type'] = new_data['type'].str.replace('Residential', 'Residential')
new_data['type'] = new_data['type'].str.replace('Industrial Area', 'Industrial')
new_data['type'] = new_data['type'].str.replace('Industrial', 'Industrial')
new_data['type'] = new_data['type'].str.replace('Industrials', 'Industrial')
new_data['type'] = new_data['type'].str.replace('Industrial Areas', 'Industrial')
new_data['type'] = new_data['type'].str.replace('Sensitive Areas', 'Others')
new_data['type'] = new_data['type'].str.replace('Sensitive Area', 'Others')
new_data['type'] = new_data['type'].str.replace('RIRUO', 'Others')
new_data['type'] = new_data['type'].str.replace('Sensitive', 'Others')
```
